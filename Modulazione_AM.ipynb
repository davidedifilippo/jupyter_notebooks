```python
import numpy as np
import matplotlib.pyplot as plt


#Portante 

Ap = 10
fp = 10

Tp = 1/fp

t = np.linspace(0, 2, num=1000000)

sp = Ap * np.cos(2*np.pi * fp * t)

plt.title('Signal')
plt.plot(t, sp, 'b')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid()
plt.show()
```

```python
#Modulante

Am = 6
fm = 1

sm = Am * np.cos(2*np.pi * fm * t)

plt.title('Signal')
plt.plot(t, sm, 'b')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid()
plt.show()
```

```python
#Modulato

k=1

smodulato = (Ap+k*sm)* np.cos(2*np.pi * fp * t)

#Andamento nel tempo

plt.title('Signal')
plt.plot(t, smodulato, 'b')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid()
plt.show()
```

```python
#Modulato

k=1

#componenti in frequenza

smodulato = sp + (Am/2) * np.cos(2*np.pi * (fp+fm) * t) + (Am/2) * np.cos(2*np.pi * (fp-fm) * t)

plt.title('Signal')
plt.plot(t, smodulato, 'b')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid()
plt.show()
```
