# **Caso di carico resistivo: Tensione e corrente sono in fase**

**Tensione:**

$$ v(t) = V_{max} sin( 2\pi f t+\phi_{v})$$

```python
import numpy as np 
import matplotlib.pyplot as plt

Vmax = 1
frequency = 10000
phi_degree = 0

# Signal period

T = 1/frequency

# Signale Root Mean Square

V_RMS = Vmax/np.sqrt(2)

# Initial phase angle in radians

phi_rad = phi_degree/180 * np.pi

t = np.linspace(0, 2*T *1000, num=1000)

v = Vmax * np.sin(2*np.pi * frequency * t + phi_rad)


plt.title('Tensione istantanea')
plt.plot(t, v, 'b')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensione [Volt]')
plt.grid()
plt.show()

```

**Corrente:**

$$ i(t) = I_{max} sin( 2\pi f t+\phi_{i})$$

```python
Imax = 1
phi_degree = 0

# Valore efficace o Root Mean Square

I_RMS = Imax/np.sqrt(2)

# Initial phase angle in radians

phi_rad = phi_degree/180 * np.pi

t = np.linspace(0, 2*T *1000, num=1000)

i = Vmax * np.sin(2*np.pi * frequency * t + phi_rad)


plt.title('Corrente istantanea')
plt.plot(t, i, color = "orange")
plt.xlabel('tempo [s]')
plt.ylabel('Corrente [Ampere]')
plt.grid()
plt.show()

```

**Potenza istantanea:**

$$ p(t) = v(t)*i(t) $$




```python
p = v*i

plt.title('Potenza istantanea')
plt.plot(t, p, color = "red")
plt.xlabel('tempo [s]')
plt.ylabel('Potenza [Watt]')
plt.grid()
plt.show()
```

Si vede che **la potenza istantanea ha frequenza doppia** rispetto alla frequenza di alimentazione mentre il suo valore medio  V_RMS * I_RMS



**La potenza media o potenza attiva**

$$P = V_{RMS} * I_{RMS} * \cos(\phi_{v}-\phi{i})$$

Essendo in quadratura di fase:

 $$P = V_{RMS} * I_{RMS} *cos(0) = V_{RMS} * I_{RMS}$$

```python
P = V_RMS*I_RMS

print("Potenza attiva: ", np.round(P,2), "Watt")
```

**L'induttore e il condensatore assorbono/cedono potenza in modo duale**

**Potenza Reattiva induttiva**

$$Q_{L} = V_{RMS} * I_{RMS} * \sin(\phi_{v}-\phi{i})$$

Essendo la tensione sfasata di 90:

$$Q_{L} = V_{RMS} * I_{RMS} * \sin(\phi_{i} + 90-\phi{i})$$

Pertanto:

$$Q_{L} = V_{RMS} * I_{RMS} * \sin( 90)$$  

Per cui:

$$Q_{L} =  V_{RMS} * I_{RMS} $$
